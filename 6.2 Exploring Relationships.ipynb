### 6. Create a categorical plot and interpret the results.

```python
# Highest price by trend

# 1. set figure size
fig = plt.figure(figsize=(20,18) )

# 2. group by the Price trend and Living Space
engine_price = df_rent.groupby('Price trend')['Living Space'].mean().sort_values().reset_index(drop = False)

# 3. Plot the graph
sns.barplot(x ='Price trend', y = 'Living Space', data = engine_price,
                 palette = 'crest')
plt.title("Average Rent Price")
plt.xticks(rotation = 90)
```

```python
# Highest viewed by engine type

# 1. set figure size
fig = plt.figure(figsize=(20,18) )

# 2. group by the Price trend and Total Rent
engine_viewed = df_rent.groupby('Price trend')['Total Rent'].mean().sort_values().reset_index(drop = False)

# 3. Plot the graph
sns.barplot(x ='Price trend', y = 'Total Rent', data = engine_viewed,
                 palette = 'crest')
plt.title("Average rent - popularity by Price trend")
plt.xticks(rotation = 90)
```

```python
df_rent.columns
```

```python
# Highest viewed by No Rooms Range

# 1. set figure size
fig = plt.figure(figsize=(18,12) )

# 2. group by No Rooms Range and Total Rent
mat_viewed = df_rent.groupby('No Rooms Range')['Total Rent'].mean().sort_values().reset_index(drop = False)

# 3. Plot the graph
sns.barplot(x ='No Rooms Range', y = 'Total Rent', data = mat_viewed,
                 palette = 'crest')
plt.title("Average Rent - popularity by No Rooms Range")
plt.xticks(rotation = 50)

# No Rooms Range most popular
```
